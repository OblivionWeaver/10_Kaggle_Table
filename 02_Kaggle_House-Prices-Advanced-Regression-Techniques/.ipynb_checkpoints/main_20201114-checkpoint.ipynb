```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Encoder
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')
sub = pd.read_csv('./data/sample_submission.csv')
```

```python
test['SalePrice'] = np.nan
train_x = train.drop('SalePrice',axis=1)
train_y = train['SalePrice']
df = pd.concat([train_x,test],ignore_index=True,sort=False)
```

```python
df
```

```python
#20 Columns
Tier2 =  ['MSSubClass','MSZoning','LotArea','LandContour','Neighborhood',
        'Condition1','Condition2','HouseStyle','OverallQual','YearBuilt',
        'YearRemodAdd','TotalBsmtSF','1stFlrSF','2ndFlrSF',
        'GrLivArea','TotRmsAbvGrd','SaleType','SaleCondition']
```

<h3>Feature Engneering</h3>

```python
df = df[Tier2]
```

#null
- MSZoning :4
- Exterior1st:1
- TotalBsmtSF:1
- SaleType:1

```python
df
```

```python
df.dropna(how='any', axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
#Label Encoding
for i in range(df.shape[1]):
    if df.iloc[:,i].dtypes == object:
        LE.fit(list(df.iloc[:,i].values))
        df.iloc[:,i] = LE.transform(list(df.iloc[:,i].values))
```

<h3>Modelling</h3>

```python
df
```

```python
train_x = df[:len(train_x)]
test_x = df[len(train):]
```

```python
test_x
```

```python
test_x
```

```python
#Convert to numpy array
X = train_x
Y = train_y
test_x =test_x
```

```python
X
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate
```

```python
select = SelectKBest(k= 15)

clf = RandomForestClassifier(random_state=10,
                             warm_start=True,
                            n_estimators = 26,
                            max_depth=6,
                            max_features='sqrt')
pipeline = make_pipeline(select,clf)
pipeline.fit(X,Y)
```

<h3>Evaluation</h3>

```python
from sklearn.metrics import mean_squared_error
cv_result = cross_validate(pipeline, X, Y, cv= 10)
print('RMSR = ', np.sqrt(cv_result['test_score']))
```

```python
# --------　採用した特徴量 ---------------
# 採用の可否状況
mask= select.get_support()

# 項目のリスト
list_col = list(df.columns[1:])

# 項目別の採用可否の一覧表
for i, j in enumerate(list_col):
    print('No'+str(i+1), j,'=',  mask[i])

# シェイプの確認
X_selected = select.transform(X)
print('X.shape={}, X_selected.shape={}'.format(X.shape, X_selected.shape))

# ----- Submit dataの作成　------- 
Id=test['Id']
predictions = pipeline.predict(test_x)
submission = pd.DataFrame({"Id": Id, "SalePrice": predictions})
submission.to_csv("submission20201115.csv", index=False)
```
